### Redução da Granularidade.

### Remove os dados repetidos, as linhas repetidas.

Função que seta como false o ultimo elemento duplicado. Depois retorna o index em data dos valores que estão == false

```python
# Os elementos com valor False são os últimos elementos da lista
# com itens repetidos
def getLastTransaction(data):
    
    sessionNums = data.sessionNo.duplicated(keep = 'last')
    
    index = [i for i in range(len(sessionNums)) if sessionNums[i] == False]
    
    
    return index
```

# **Pré-processamento**

### Cria uma lista com todas as variáveis objetos(string) do banco de dados.

```python
objects = x_train.select_dtypes(include = 'object').columns.tolist()
```

### Separa as variáveis númericas das que devem ser mantidas como string e remove da lista a coluna de identificação 'customerNo'.

```python
strings = ['onlineStatus', 'availability']
numeric = deepcopy(objects)
numeric.remove('onlineStatus')
numeric.remove('availability')
objects.remove('customerNo')
```

### Substitui os valores ausentes por zero quando for um novo cliente.

```python
# Substitui '?' por 0, quando 'customerNo' == '?'
def zeroNewClient(data):
    labels = ['maxVal','customerScore', 'accountLifetime', 'payments', 'age', 'address', 'lastOrder']
    mask = data['customerNo'] == '?'
    for i in labels:
        data.loc[mask, i] = 0
```

```python
zeroNewClient(x_train)
zeroNewClient(x_test)
```

### Substitui os valores ausentes. Substitui '?' por np.Nan.

```python
def replaceMissing(data, labels):
    for i in labels:
        data.loc[:, i] = data.loc[:, i].apply(lambda v: np.NaN if v == '?' else v)
```

```python
replaceMissing(x_train, objects)
replaceMissing(x_test, objects)
```

```python
def replaceMissingString (data, labels):
    for i in labels:
        data[i].fillna('ausente', inplace = True )
```

```python
replaceMissingString(x_train, strings)
replaceMissingString(x_test, strings)
```

### Etapa de criação de novas variáveis para se recuperar informações perdidas durante a fase de redução de granularidade.



*   cMean: Média dos valores de todos os produtos clicados pelo usuário
*   bMean: Média dos valores de todos os produtos inseridos ao carrinho pelo usuário
*   countDiffCB: Diferença entre o máximo de todos os itens clicados pelos itens inseridos no carrinho
*   qtTransactions: Quantidade de transações realizadas pelo usuário
*   bStepMean: Média entre todos os bStep de cada usuário



```python
#criando as variáveis
def createVariableMean(data, obj_labels, num_labels):
    replaceMissing(data, obj_labels)
    toNumeric(data, num_labels)
    
    meanC = data.groupby('sessionNo').cSumPrice.mean()
    meanB = data.groupby('sessionNo').bSumPrice.mean()
    cMax = train_copy.groupby('sessionNo').cCount.max()
    bMax = train_copy.groupby('sessionNo').bCount.max()
    countDiffCB = cMax - bMax
    qtTransactions = train_copy.groupby('sessionNo').sessionNo.count()
    bStepMean = train_copy.groupby('sessionNo').bStep.mean()
    
    return meanC, meanB, countDiffCB, qtTransactions, bStepMean
```

### Criando as novas variáveis para os dados de treinamento

```python
meanC, meanB, cDiffB, qtTrans, bsm = createVariableMean(train_copy, objects, numeric)
x_train = x_train.assign(cMean = meanC, bMean = meanB, countDiffCB = cDiffB, qtTransactions = qtTrans, bStepMean = bsm)
x_train.shape
```

### Criando as novas variáveis para os dados de test

```python
meanC, meanB, cDiffB, qtTrans, bsm = createVariableMean(test_copy, objects, numeric)
x_test = x_test.assign(cMean = meanC, bMean = meanB, countDiffCB = cDiffB, qtTransactions = qtTrans, bStepMean = bsm)
x_test.shape
```

### Insere as novas variáveis na lista de variáveis numéricas.

```python
numeric.extend(['cMean', 'bMean', 'countDiffCB', 'qtTransactions', 'bStepMean'])
```

### Funções responsáveis por substituir os valores numéricos ausentes por um valor estatístico específico.

```python
# Substitui pelo valor Máximo
def replaceWithMax(data, labels):
    for i in labels:
        maximum = data[i].max(axis = 0)
        data.loc[:, i].fillna(value = maximum, inplace = True)

# Substitui pelo valor Mínimo
def replaceWithMin(data, labels):
    for i in labels:
        maximum = data[i].min(axis = 0)
        data.loc[:, i].fillna(value = maximum, inplace = True)

# Substitui pelo valor Média
def replaceWithMean(data, labels):
    for i in labels:
        maximum = data[i].mean(axis = 0)
        data.loc[:, i].fillna(value = maximum, inplace = True)

# Substitui pelo valor Mediana
def replaceWithMedian(data, labels):
    for i in labels:
        maximum = data[i].median(axis = 0)
        data.loc[:, i].fillna(value = maximum, inplace = True)
```

### Substititui os valores ausentes pela estatística

```python
type_stat = 'max'
replaceWithMax(x_train, numeric)
replaceWithMax(x_test, numeric)
```
